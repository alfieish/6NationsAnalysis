```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = [9, 9]
```

```python
path = r'C:\Users\Maurice\Desktop\Python\SixNations\fivePlusPlayers01.csv'
playerDF = pd.read_csv(path,encoding = "ISO-8859-1")
playerDF.drop(['Unnamed: 0'], axis=1, inplace=True)
print(playerDF.sample(5))
print(playerDF.describe())
```

```python
positionDFs=[]
for p in playerDF['Position'].unique():
        DF = playerDF.loc[playerDF['Position'] == p]
        DF.name = p
        positionDFs.append(DF)

print(len(positionDFs))
print(positionDFs[0].sample(5))
print(positionDFs[0].columns)
```

```python
for pDF in positionDFs:
    individDF = pDF.groupby('Name').mean().sort_values('ITScore', ascending=False)
    individDF.drop('Try', axis=1, inplace=True)
    countDF = pDF.groupby('Name').count()['Try']
    individDF = pd.concat([individDF, countDF], axis=1).sort_values('ITScore', ascending=False)
    individDF.rename(columns={'Try': 'Appearances'}, inplace=True)
    individDF = individDF.loc[individDF['Appearances'] > 5]
    print(pDF.name)
    print(individDF[[ 'Appearances', 'Points', 'Try Assist', 'Passes', 'Offloads', 'Clean Breaks', 'Tackles','Missed Tackles', 'Meters Run', 'Defenders Beaten', 'Net Points', 'ITScore', 'ITRolling']].head(10))
    print()
    individDF.to_csv('position-'+pDF.name+'.csv')
    
```
